```
import numpy as np
from sklearn.datasets import make_classification

# Generate a random n-class classification problem.
X, y = make_classification(n_samples=500, n_features=2, n_redundant=0, n_informative=1, n_clusters_per_class=1)
X = X.T
# np.newaxis comes in very handy when you want to explicitly convert an 1D array to either a row vector or a column vector.
y = y[:,np.newaxis].T

assert (X.shape == (2,500))
assert (y.shape == (1,500))
```

```
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('white')
sns.scatterplot(X[0,:],X[1,:], hue=y.reshape(-1));
```
